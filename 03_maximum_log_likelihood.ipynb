```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
y = np.random.poisson(7, size=1000)
sns.distplot(y)
```

```python
def poisson_likelihood(x,l):
    return np.exp(-l)*(l**x) / (np.arange(x)+1).prod()

poisson_log_likelihood = lambda x, lam: -lam + np.log(lam**x) - np.log(np.arange(x)+1).sum()

poisson_likelihood(10,6)
poisson_log_likelihood(10,6)
```

```python
l=7
np.sum([poisson_log_likelihood(yv, l) for yv in y])
```

```python
lambdas = range(3,12,1)
# y= np.arange(15)
lls =  []
for l in lambdas:
    lls.append(np.sum([poisson_log_likelihood(yv, l) for yv in y]))
```

```python
# lls.index(np.max(lls))
lls
```

```python
# print(lls.index(lls.max()))
plt.plot(lambdas,lls)
plt.xlabel('$\lambda$')
# plt.ylabel('L($\lambda$|x={0})'.format(y));
plt.ylabel('L($\lambda$|y)')
```
